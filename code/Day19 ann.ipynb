```python
class Layer_Dense:
    def __init__(self,m,n_l):
        '''m is n_inputs,i.e. number of features, n_l is n_neurons in the layer'''
        self.w=0.01*np.random.randn(m,n_l) #no transpose needed
        # print("created " , self.w.shape)
        self.bias=np.zeros((1,n_l))
        # self.output=None
        pass

    def forward(self,inputs):
        # pass
        w=self.w
        bias=self.bias
        # print(w.shape, X.shape)
        self.output= inputs@w + bias
```
